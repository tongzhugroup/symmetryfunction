```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def FC(Rij):
    if Rij<=Rc:
        return 0.5*(np.cos(np.pi*(Rij/Rc))+1)
    else:
        return 0
```

```python
#G1
RcL=[3.0,5.0,7.0,9.0,11.3]
for Rc in RcL:
    X=np.linspace(0,14,100)
    Y=[FC(i) for i in X]
    plt.plot(X,Y)
    plt.ylim(0,1)
    plt.xlim(0,14)
    plt.legend(["$R_c$=3.0 Bohr","$R_c$=5.0 Bohr","$R_c$=7.0 Bohr","$R_c$=9.0 Bohr","$R_c$=11.3 Bohr"])
```

```python
def G2(Rij):
    return np.exp(-eta*np.square(Rij-Rs))*FC(Rij)
Rc=11.3
Rs=0
etaL=[5.00,1.00,0.40,0.20,0.10,0.06,0.03,0.01]
for eta in etaL:
    X=np.linspace(0,14,500)
    Y=[G2(i) for i in X]
    plt.plot(X,Y)
    plt.ylim(0,1)
    plt.xlim(0,14)
    plt.legend(["$\eta=5.00$ $Bohr^{-2}$","$\eta=1.00$ $Bohr^{-2}$","$\eta=0.40$ $Bohr^{-2}$","$\eta=0.20$ $Bohr^{-2}$","$\eta=0.10$ $Bohr^{-2}$","$\eta=0.06$ $Bohr^{-2}$","$\eta=0.03$ $Bohr^{-2}$","$\eta=0.01$ $Bohr^{-2}$"])
    
```
